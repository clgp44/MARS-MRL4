```python
import matplotlib.pyplot as plt
import pandas as pd
import math

from neo4j import GraphDatabase

fileName = "drill_log - test#10 - H22"

df = pd.read_csv("%s.csv" %fileName,
                  dtype = {'Drilling UID': str, 'Time (µs)': int, 'T°': int, 'RPM': int, 'Spindle Amps (%)': float, 'Spindle Cmd (%)': float, 'Feed Pos. (µm)': int, 'Feed Amps (%)': float, 'Feed Cmd (%)': float, 'Clamp Pos. (µm)': int, 'Clamp Amps (%)': float, 'Clamp Cmd (%)': float})

             
df.head()
```

```python
# Data for plotting
t = df['Time (µs)']/1000
s = df['Feed Amps (%)']*55/100

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Time (ms)', ylabel='Feed Amps (A)',
       title=fileName)
ax.grid()

fig.savefig("%s - Feed Amps.png" %fileName)
plt.show()
```

```python
# Data for plotting
t = df['Time (µs)']/1000
s = df['Feed Cmd (%)']*24/100

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Time (ms)', ylabel='Feed Cmd (V)',
       title=fileName)
ax.grid()

fig.savefig("%s - Feed Cmd.png" %fileName)
plt.show()
```

```python
# Data for plotting
t = df['Time (µs)']/1000
s = df['Feed Pos. (µm)']

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Time (ms)', ylabel='Feed Pos. (µm)',
       title=fileName)
ax.grid()

fig.savefig("%s - Feed Pos.png" %fileName)
plt.show()
```

```python
# Data for plotting
t = df.loc[200:250, 'Time (µs)']/1000
s1 = df.loc[200:250, 'Feed Cmd (%)']*24/100
s2 = df.loc[200:250, 'Feed Amps (%)']*55/100
s3 = df.loc[200:250, 'Feed Pos. (µm)']

fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharey=False)

# axs[0].grid(True)
# axs[0].set(xlabel='Time (ms)', ylabel='Feed Cmd (V)')
# axs[0].plot(t,s1)

axs[0].grid(True)
axs[0].set(xlabel='Time (ms)', ylabel='Feed Amps (A)')
axs[0].plot(t,s2)

axs[1].grid(True)
axs[1].set(xlabel='Time (ms)', ylabel='Feed Pos. (µm)')
axs[1].plot(t,s3)


fig.suptitle('%s - Spindle' %fileName)
fig.savefig("%s - Amp - Feed.png" %fileName)
```
